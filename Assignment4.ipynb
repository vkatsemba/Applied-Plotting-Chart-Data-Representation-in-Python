```python
#Import modules
import math
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#Read and observe the Gas Consumption data
gas_df = pd.read_csv('GasConsumption.csv', encoding='cp1252', skiprows=4)
```

```python
#Filter the rows with data for Belarus.
gas_df = gas_df.loc[(gas_df['Country Name'] == "Belarus")]
```

```python
#Remove columns with data before 1992 and from 2015 to the present as there is no data for those time periods.
gas_df.reset_index(drop=True, inplace=True)
years_without_data = ['2015', '2016', '2017', '2018','2019', 'Unnamed: 64']
gas_df.drop(years_without_data, axis=1, inplace=True)
gas_df.drop(gas_df.iloc[:, 1:36], inplace = True, axis = 1)
gas_df
```

```python
#Read and observe the Liquid Consumption data
liquid_df = pd.read_csv('LiquidConsumption.csv', encoding='cp1252', skiprows=4)
```

```python
#Filter the rows with data for Belarus.
liquid_df = liquid_df.loc[(liquid_df['Country Name'] == "Belarus")]
```

```python
#Remove columns with data before 1992 and from 2015 to the present as there is no data for those time periods.
liquid_df.reset_index(drop=True, inplace=True)
years_without_data = ['2015', '2016', '2017', '2018','2019', 'Unnamed: 64']
liquid_df.drop(years_without_data, axis=1, inplace=True)
liquid_df.drop(liquid_df.iloc[:, 1:36], inplace = True, axis = 1)
liquid_df
```

```python
#Read and observe the Solid Consumption data
solid_df = pd.read_csv('SolidConsumption.csv', encoding='cp1252', skiprows=4)
```

```python
#Filter the rows with data for Belarus.
solid_df = solid_df.loc[(solid_df['Country Name'] == "Belarus")]
```

```python
#Remove columns with data before 1992 and from 2015 to the present as there is no data for those time periods.
solid_df.reset_index(drop=True, inplace=True)
years_without_data = ['2015', '2016', '2017', '2018','2019', 'Unnamed: 64']
solid_df.drop(years_without_data, axis=1, inplace=True)
solid_df.drop(solid_df.iloc[:, 1:36], inplace = True, axis = 1)
solid_df
```

```python
#Merge the three rows into one dataframe
df = pd.concat([gas_df, liquid_df, solid_df], axis = 0)
df
```

```python
#Change the column "Country Name" to "Energy State" and replace "Belarus" with Gas, Liquid and Solid respectively.
df = df.rename(columns = {'Country Name': 'Energy State'})
df.reset_index(drop=True, inplace=True)
df.loc[0, "Energy State"] = 'Gas'
df.loc[1, "Energy State"] = 'Liquid'
df.loc[2, "Energy State"] = 'Solid'
df
```

```python
#Transpose the data
df = df.transpose()
df.columns = df.iloc[0]
df = df[1:]
df
```

```python
#Generate a plot
df.plot(colormap='Set1', figsize=(10,5), lw=5);
plt.xlabel('Years')
plt.ylabel('CO2 Emissions(kilotons)')
plt.title('Belarus Carbon Dioxide Emissions from Gaseous, Liquid and Solid Fuel Consumption 1992-2014')
plt.show()
```
